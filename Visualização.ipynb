# **Gráfico de dispersão**

Breve resumo:

O gráfico de dispersão é um tipo de gráfico que mostra a relação entre duas variáveis. Ele é criado plotando os pontos de dados em um plano cartesiano, com a variável independente no eixo x e a variável dependente no eixo y.

Quando utilizar:

- O gráfico de dispersão deve ser utilizado sempre que se deseja visualizar a relação entre duas variáveis. Ele é uma ferramenta útil para identificar padrões e tendências, como correlação, causalidade e outliers.

Exemplos:

- Um pesquisador pode usar um gráfico de dispersão para visualizar a relação entre o nível de escolaridade e o salário.

```python
# Criando o gráfico
sns.relplot(data=df, x = 'age', y = 'final-weight',
            hue = 'income', style = 'sex', size = 'education-num');
```

Para entendermos melhor esse gráfico, vamos observar para que servem esses parâmetros passados na criação do mesmo:

- data: O conjunto de dados a ser usado para criar o gráfico.
- x: A variável independente do gráfico.
- y: A variável dependente do gráfico.
- hue: A variável categórica para colorir os pontos do gráfico.
- style: A variável categórica para estilizar os pontos do gráfico.
- size: A variável contínua para controlar o tamanho dos pontos do gráfico.

Então é possível observar que quanto maior é o tamanho do ponto, maior é o grau de escolaridade desse registro presente na base. Ao mesmo tempo, estamos utilizando uma coloração laranja para rendas maiores que 50 mil e uma coloração azul para rendas menores que 50 mil. Caso o registro seja do sexo feminino, o seu formato será um X. Caso contrário, será um círculo. Com esses pontos em mente, podemos chegar às seguintes conclusões:

Pessoas com rendas superiores a 50 mil normalmente têm mais de 30 anos. Em sua grande maioria, são do sexo masculino e, aparentemente, quanto maior o nível de escolaridade, maior a renda do indivíduo.

Esse gráfico está exibindo bastante informações de uma vez só. Talvez essa abordagem não seja a mais indicada, porém o intuito da criação dele, nesse caso, é a exemplificação dos parâmetros de criação do mesmo.

# **Gráfico de barras**

Breve resumo:

O gráfico de barras é um tipo de gráfico que usa barras para representar valores categóricos. As barras são geralmente dispostas horizontalmente ou verticalmente.

Quando utilizar:

- O gráfico de barras deve ser utilizado sempre que se deseja visualizar dados categóricos. Ele é uma ferramenta útil para comparar valores entre diferentes categorias.

Exemplos:

- Um pesquisador pode usar um gráfico de barras para visualizar a distribuição de idades de uma população.
